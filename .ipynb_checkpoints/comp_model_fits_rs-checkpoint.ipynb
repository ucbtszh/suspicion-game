```python
# get all DF filenames from data/fr folder

def data_to_df(files: list):
    ''' provide list of feather data file pathnames '''
    dfs = {}  # features of interest
    for path in files:
        try:
            df = pd.read_feather("./data/rs/"+path)
            uuid = re.search('(?<=rbatch_).*(?=\.feather)', path)
            if (uuid is None) | (len(uuid[0]) < 24):
                continue
#             print(uuid[0])
            dfs[uuid[0]] = df
        except PermissionError:
            continue
    return dfs
```

```python
param_search_space = [Real(0, 1, name='prior'), Real(0, 1, name='alpha1'), Real(0, 1, name='alpha2'), Real(0, 1, name='weight')]
```

```python
def combi_model_fit(df_long, uuids, param_search_space, stat1: str, stat2: str):
    results = {}
    
    for uuid in uuids:
        trials = df_long[df_long["PID"]==uuid]
        hrs = trials['suspicion_rating'].values
        result = skopt_fit_weighted_model_single_response(hrs, trials, param_search_space, stat1, stat2)
        results[uuid] = result

    return results
```

```python
def objective_multi_params(params, response, trials, stat1: str, stat2: str, stat3: str):
    pred = params[0] + params[1] * trials[stat1] + params[2] * trials[stat2] + params[3] * trials[stat3]
    ss_res = np.sum((response-pred)**2)
    return ss_res
```

```python
def skopt_fit_multi_param(response, trials, param_search_space, stat1: str, stat2: str, stat3: str):
    ss_tot = np.sum((response - np.mean(response)) ** 2)
    gp_result = gp_minimize(partial(objective_multi_params, response=response, trials=trials, stat1=stat1, stat2=stat2, stat3=stat3),
                            param_search_space, random_state=42)
    optimal_ss_res = gp_result.fun
    
    r2 = 1 - np.divide(optimal_ss_res, ss_tot)

    print("Best parameter estimates: prior =", gp_result.x[0], "alpha 1 =", gp_result.x[1], "alpha 2 =", gp_result.x[2],
         "alpha 3 =", gp_result.x[3])
    print("R2:", r2)

    pred = gp_result.x[0] + gp_result.x[1] * trials[stat1] + gp_result.x[2] * trials[stat2] + gp_result.x[3] * trials[stat3]

    mse = mean_squared_error(response, pred)
    bic = calculate_bic(len(response), mse, len(param_search_space))
    aic = calculate_aic(len(response), mse, len(param_search_space))

    print("BIC:", bic)
    print("AIC:", aic)
    print("=" * 100)
    
    return {"ss_total": ss_tot, "est_prior": gp_result.x[0], "est_alpha1": gp_result.x[1], "est_alpha2": gp_result.x[2], "est_alpha3": gp_result.x[3], "mean_squared_error": mse, "R2": r2, "BIC": bic, "AIC": aic}
```

```python
def combi_model_fit_edt(df_long, uuids, param_search_space, stat1: str, stat2: str, stat3: str):
    results = {}
    
    for i, uuid in enumerate(uuids):
        print(i, uuid)
        df = df_long[df_long['PID']==uuid]
        hrs = df['suspicion_rating'].values
        result = skopt_fit_multi_param(hrs, df, param_search_space, stat1, stat2, stat3)
        results[uuid] = result

    return results
```

## Compute subjects' lie detection accuracy

```python
from scipy.stats import norm
```

```python
lie_detection = {}

for i, uuid in enumerate(uuids):
    idf = df[df["PID"]==uuid]
    detected_deception = [1 if float(r) > 0.5 else 0 for r in idf['suspicion_rating'].values]
    
    tp = [1 if (x==1) & (y==1) else 0 for x, y in zip(idf['pp_lied'], detected_deception)]
    tp = sum(tp)
    
    fp = [1 if (x==0) & (y==1) else 0 for x, y in zip(idf['pp_lied'], detected_deception)]
    fp = sum(fp)
    
    tn = [1 if (x==0) & (y==0) else 0 for x, y in zip(idf['pp_lied'], detected_deception)]
    tn = sum(tn)
    
    fn = [1 if (x==1) & (y==0) else 0 for x, y in zip(idf['pp_lied'], detected_deception)]
    fn = sum(fn)
    
    accuracy = (tp + tn)/len(idf)
    
    if tp+fp == 0:
        precision = 0
    else:
        precision = tp/(tp+fp)
    
    if tp+fn == 0:
        recall=0
    else:
        recall = tp/(tp+fn)
        
    if tn+fp == 0:
        specifity = 0
    else:
        specifity = fp/(fp+tn)
        
    hitrate = tp/sum(idf["pp_lied"]==1)
    farate = fp/sum(idf["pp_lied"]==0)
    
    assert(0<=hitrate<1)
    assert(0<=farate<1)
    
    dprime = norm.cdf(hitrate) - norm.cdf(farate)
#     print(dprime)
    
    f1 = tp/(tp+0.5*(fp+fn))
    
    lie_detection[uuid] = {"n_subjective_detection": sum(detected_deception), 
                           "tp": tp, 
                           "fp": fp, 
                           "tn": tn, 
                           "fn": fn, 
                           "precision": precision, 
                           "recall": recall, 
                           "accuracy": accuracy, 
                           "specificity": specifity,
                           "dprime": dprime,
                           "f1": f1}
```

```python
with open("./data/rs/lie_detection_accuracy.json", "w", encoding="utf-8") as f:
    json.dump(lie_detection, f)
```
