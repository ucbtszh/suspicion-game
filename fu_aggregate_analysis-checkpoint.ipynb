```python
lie_detection = {}

for i, uuid in enumerate(uuids):
    sr = suspicion_ratings[uuid]
    detected_deception = [1 if float(r) > 0.5 else 0 for r in sr]
    
    dd_b1 = sum(detected_deception[:30])
    dd_b2 = sum(detected_deception[30:60])
    dd_b3 = sum(detected_deception[60:])
    
    tp_b1 = [1 if (x==1) & (y==1) else 0 for x, y in zip(pp_lies[uuid][:30], detected_deception[:30])]
    tp_b1 = sum(tp_b1)
#     print(tp_b1)
    
    tp_b2 = [1 if (x==1) & (y==1) else 0 for x, y in zip(pp_lies[uuid][30:60], detected_deception[30:60])]
    tp_b2 = sum(tp_b2)
#     print(tp_b2)
    
    tp_b3 = [1 if (x==1) & (y==1) else 0 for x, y in zip(pp_lies[uuid][60:], detected_deception[60:])]
    tp_b3 = sum(tp_b3)
#     print(tp_b3)

    fp_b1 = [1 if (x==0) & (y==1) else 0 for x, y in zip(pp_lies[uuid][:30], detected_deception[:30])]
    fp_b1 = sum(fp_b1)
#     print(fp_b1)
    
    fp_b2 = [1 if (x==0) & (y==1) else 0 for x, y in zip(pp_lies[uuid][30:60], detected_deception[30:60])]
    fp_b2 = sum(fp_b2)
#     print(fp_b2)
    
    fp_b3 = [1 if (x==0) & (y==1) else 0 for x, y in zip(pp_lies[uuid][60:], detected_deception[60:])]
    fp_b3 = sum(fp_b3)
#     print(fp_b3)
    
    try:
        hitrate_b1 = tp_b1/dd_b1 
    except ZeroDivisionError:
        hitrate_b1 = 0
        
    try:
        hitrate_b2 = tp_b2/dd_b2
    except ZeroDivisionError:
        hitrate_b2 = 0
    
    try:
        hitrate_b3 = tp_b3/dd_b3
    except ZeroDivisionError:
        hitrate_b3 = 0  
        
    try:
        farate_b1 = fp_b1/(30-dd_b1)
    except ZeroDivisionError:
        farate_b1 = 0
        
    try:
        farate_b2 = fp_b2/(30-dd_b2)
    except ZeroDivisionError:
        farate_b2 = 0
    
    try:
        farate_b3 = fp_b3/(30-dd_b3)
    except ZeroDivisionError:
        farate_b3 = 0
    
    print(fp_b1, fp_b2, fp_b3)
    print(hitrate_b1, hitrate_b2, hitrate_b3, farate_b1, farate_b2, farate_b3)
    
    # farate can be > 1 if a participant was "overly suspicious" vs the actual amount of lies
    
    dprime_b1 = norm.cdf(hitrate_b1) - norm.cdf(farate_b1)
    dprime_b2 = norm.cdf(hitrate_b2) - norm.cdf(farate_b2)
    dprime_b3 = norm.cdf(hitrate_b3) - norm.cdf(farate_b3)
    
    lie_detection[uuid] = {"n_subjective_detection": sum(detected_deception), 
                           "block1_tp": tp_b1, 
                           "block1_fp": fp_b1,
                           "block1_dprime": dprime_b1,
                            "block2_tp": tp_b2, 
                           "block2_fp": fp_b2,
                           "block2_dprime": dprime_b2,
                            "block3_tp": tp_b3, 
                           "block3_fp": fp_b3,
                           "block3_dprime": dprime_b3,
#                             "counterbalance": order[uuid],
#                           "age": age[uuid],
#                            "edlev": edlev[uuid],
#                            "gender": gender[uuid],
                           "block1_lie_prop": lie_prop_b1[uuid],
                           "block2_lie_prop": lie_prop_b2[uuid],
                           "block3_lie_prop": lie_prop_b3[uuid],
#                            "crt": crt_responses[uuid]
                          }
```

## CRT scores

```python
# demographics survey query reference
refs_crt = []

for uuid in uuids:
    refs_crt.append(subjects_ref.document(uuid).collection(u'responses').document('CRT').get())
```

```python
len(refs_crt)
```

```python
crt_responses = {}

for i, refs in enumerate(refs_crt):
    print(refs.to_dict())
    score = 0
    
    crt1 = refs.to_dict()['crt1']
    if str(5) in crt1:
        score += 1
    
    crt2 = refs.to_dict()['crt2']
    if crt2 == '5':
        score += 1
    
    crt3 = refs.to_dict()['crt3']
    if crt3 == '47':
        score += 1
    
    crt4 = refs.to_dict()['crt4']
    if ('2' in crt4) | ('second' in crt4.lower()):
         score += 1
    crt_responses[uuids[i]] = score
```
